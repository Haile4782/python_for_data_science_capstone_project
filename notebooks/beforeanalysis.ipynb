## Addiction population uncleaned data pre analysis

```python
# Data Exploration for Addiction Population Data

import pandas as pd

df = pd.read_csv("../data/raw/addiction_population_data.csv")
df.head()
```

```python
# Check dataset shape & info

print("Shape:", df.shape)
df.info()
```

```python
# Check missing (null) values

df.isnull().sum().sort_values(ascending=False)
```

```python
# To see percentage:
(df.isnull().mean() * 100).sort_values(ascending=False)
```

```python
# Check duplicates

df.duplicated().sum()
```

```python
# Check empty strings

(df == "").sum()
```

```python
# Check outliers (numerical columns)

df.describe()
```

```python
# IQR method

numeric_cols = df.select_dtypes(include="number").columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: {outliers} outliers")
```

```python
# Check value consistency (categorical)

for col in df.select_dtypes(include="object").columns:
    print(col, df[col].nunique())
    print(df[col].unique())
    print("------")
```

```python
# Range checks

df["age"].min(), df["age"].max()
df["sleep_hours"].describe()
df["bmi"].describe()
```

```python
# Logical checks

df[df["age_started_smoking"] > df["age"]]
```

```python
df[df["age_started_drinking"] > df["age"]]
```
